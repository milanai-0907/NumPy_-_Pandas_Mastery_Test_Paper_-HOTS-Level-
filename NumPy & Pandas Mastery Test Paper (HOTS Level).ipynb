*1.  Financial Data Normalization & Correlation Analysis: Generate a NumPy array representing
daily stock returns for 5 stocks over 100 days (random values between -0.05 and 0.05). a)
Standardize each stock's returns (column-wise). b) Compute the correlation matrix using
NumPy. c) Identify which pair of stocks has the highest positive correlation. d) Explain what
high correlation implies in portfolio diversification.*

```python
import numpy as np
import pandas as pd

returns=np.random.uniform(-0.5,0.5,size=(100,5))
returns

#a.standardizing stock's returns:
standardized_returns=(returns-np.mean(returns,axis=0)/np.std(returns,axis=0))
standardized_returns

#b.computing correlation matrix:
corr_matrix=np.corrcoef(standardized_returns,rowvar=False)
corr_matrix

#c. identifying the pair of stocks that has the highest positive correlation:
np.fill_diagonal(corr_matrix, -np.inf) #to ignore self-correlation
max_corr_pair=np.unravel_index(np.argmax(corr_matrix),corr_matrix.shape)
max_corr_pair


```

d. following are the some key points that the high correlation implies in portfolio diversification:

-->High positive correlation (≈ +1);

- Stocks move together
- Poor diversification benefit

-->Low or negative correlation (≈ 0 or -1);
- Stocks move independently or oppositely
- Better diversification, lower overall risk


 2)  Advanced Data Cleaning & Transformation: Create a Pandas DataFrame of 50 employees
with columns: EmpID, Department (IT, HR, Sales), Salary (random), JoiningDate (random
dates in 2022–2024). a) Introduce at least 5 missing values in Salary. b) Replace missing
Salary values using department-wise median. c) Create a column 'TenureDays' (differencebetween today and JoiningDate). d) Rank employees within each department based on Salary.

```python
import numpy as np
import pandas as pd

# Fix: Convert DatetimeIndex to Series to use .sample()
df = pd.DataFrame({
    "EmployeeID": np.arange(1, 51),
    "Department": np.random.choice(["IT", "HR", "Sales"], size=50),
    "Salary": np.random.randint(20000, 50000, size=50),
    "JoiningDate": pd.date_range(start="2022-01-01", end="2024-12-31", freq="D").to_series().sample(50).values
})
df

#a.introducing 5 missing values in salaray:
df.loc[np.random.choice(df.index, 5), "Salary"] = np.nan
df

#b.replacing missing salary values using department-wise median:
df["Salary"] = df.groupby("Department")["Salary"].transform(lambda x: x.fillna(x.median()))
df

#c.creating a column 'TenureDays':
df["TenureDays"] = (pd.to_datetime("today") - df["JoiningDate"]).dt.days
df

#d rankning employees within each department based on salary:
df["SalaryRank"] = df.groupby("Department")["Salary"].rank(ascending=False)
df

```

3) Sales Analytics with Grouping & Merging: Create two DataFrames: Customers (CustomerID,
Region) and Orders (OrderID, CustomerID, OrderAmount, OrderDate). a) Merge both datasets
appropriately. b) Compute total and average sales per region. c) Identify top 3 customers by
total purchase amount. d) Compute monthly revenue trends using datetime features.

```python
import numpy as np
import pandas as pd

Customers=pd.DataFrame({
   "CostumerID":np.arange(1,11),
   "Region":np.random.choice(["East","West","South"],size=10),

})

Orders=pd.DataFrame({
   "OrderID":np.arange(1,11),
   "CostumerID":np.random.choice(np.arange(1,11),size=10),
   "OrderAmount":np.random.randint(100,1000,size=10),
   "OrderDate":pd.date_range(start="2025-12-01",end="2026-02-17",freq="D").to_series().sample(10).values
})

#a.merging the both database:
merged_df=pd.merge(Customers,Orders,on="CostumerID")
merged_df

#b. computing total and average sales per region:
total_sales_per_region=merged_df.groupby("Region")["OrderAmount"].sum()
total_sales_per_region
average_sales_per_region=merged_df.groupby("Region")["OrderAmount"].mean()
average_sales_per_region

#c.identifing top three customers by total purchase amount:
top_customers=merged_df.groupby("CostumerID")["OrderAmount"].sum().nlargest(3)
top_customers

#d.computing monthly revenue trends using datetime features:
merged_df["Month"]=merged_df["OrderDate"].dt.month
merged_df["Month"]




```

 4. Matrix-Based Linear Regression from Scratch: Using NumPy only, create synthetic dataset
with 200 samples and 3 features. a) Add bias column manually. b) Compute regression weights
using Normal Equation: (X^T X)^-1 X^T y c) Predict target values. d) Compute Mean Squared
Error. e) Explain why matrix inversion can be numerically unstable.

```python
import numpy as np
import pandas as pd


#Reproducibility
np.random.seed(0)

#number of samples and features
n_samples=200
n_features=3

#generate synthetic feature matrix
X=np.random.rand(n_samples,n_features)

#true parameters (unknown in real life)
true_weights=np.array([2.0,-1.5,0.7])
true_bias=3.0

#generate target variables with noise
y = X @ true_weights + true_bias + np.random.randn(n_samples)*0.5

# a) Add bias column (column of ones)
X_bias = np.hstack((np.ones((n_samples, 1)), X))

# b) Compute regression weights using Normal Equation
theta = np.linalg.inv(X_bias.T @ X_bias) @ X_bias.T @ y

# c) Predict target values
y_pred = X_bias @ theta

# d) Compute Mean Squared Error
mse = np.mean((y - y_pred) ** 2)

print("Learned parameters (bias + weights):")
print(theta)

print("\nMean Squared Error:")
print(mse)





```

 (e) Why Matrix Inversion Is Numerically Unstable

Matrix inversion can be unstable because:

1. Multicollinearity

  - Features are highly correlated

2. Small eigenvalues

- Inversion amplifies numerical errors

3. Floating-point precision limits

- Rounding errors accumulate

4. Poorly scaled features

- Leads to ill-conditioned matrices


5) End-to-End Mini Data Pipeline: Simulate a dataset with 500 rows including: Date,
ProductCategory, UnitsSold, Price. a) Compute total revenue per category. b) Identify category
with highest revenue growth month-over-month. c) Normalize UnitsSold using Min-Max scaling.
d) Detect outliers in Price using Z-score method. e) Provide 3 business insights from the
analysis.

```python
import pandas as pd
import numpy as np

np.random.seed(42)

n = 500

df = pd.DataFrame({
    "Date": pd.to_datetime(
        np.random.choice(pd.date_range("2024-01-01", periods=180), n)
    ),
    "ProductCategory": np.random.choice(
        ["Electronics", "Clothing", "Groceries", "Furniture", "Books"], n
    ),
    "UnitsSold": np.random.randint(1, 50, n),
    "Price": np.round(np.random.uniform(5, 500, n), 2)
})

# Revenue column
df["Revenue"] = df["UnitsSold"] * df["Price"]

df.head()

#a.total revenue per category:
revenue_per_category = df.groupby("ProductCategory")["Revenue"].sum()
print(revenue_per_category)

#b.category with highest revenue growth month-over-month:
# Extract month
df["Month"] = df["Date"].dt.to_period("M")

monthly_revenue = (
    df.groupby(["Month", "ProductCategory"])["Revenue"]
      .sum()
      .reset_index()
)

# Compute MoM growth
monthly_revenue["MoM_Growth"] = (
    monthly_revenue.groupby("ProductCategory")["Revenue"]
    .pct_change()
)

# Identify category with highest growth
top_growth_category = monthly_revenue.loc[
    monthly_revenue["MoM_Growth"].idxmax()
]["ProductCategory"]

print("Highest MoM growth category:", top_growth_category)

#c.Normalize UnitsSold using Min-Max scaling:
df["UnitsSold_Normalized"] = (
    (df["UnitsSold"] - df["UnitsSold"].min()) /
    (df["UnitsSold"].max() - df["UnitsSold"].min())
)

#d. outliers in Price using Z-score method:
price_mean = df["Price"].mean()
price_std = df["Price"].std()

df["Price_Zscore"] = (df["Price"] - price_mean) / price_std

# Outliers: |Z| > 3
outliers = df[np.abs(df["Price_Zscore"]) > 3]

print("Number of price outliers:", len(outliers))
```

e)  Business Insights
- Insight 1: Revenue Concentration

   A small number of product categories contribute the majority of total revenue, indicating where marketing and inventory investment should be focused.

- Insight 2: Growth vs Volume

   The category with the highest month-over-month growth is not always the highest-revenue category, suggesting emerging products with strong future potential.

- Insight 3: Pricing Anomalies
  
  Price outliers detected via Z-score may indicate premium products, pricing errors, or data quality issues each requiring different business actions.
